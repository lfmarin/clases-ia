# Lectura de imágenes y modelos de color


```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
img = cv2.imread("./files/bananos.jpg")#nos lee el formato en BGR (modelo de color)
plt.imshow(img)#plotea en RGB
type(img)
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
img2
#
```

```python
image= cv2.cvtColor(img2, cv2.COLOR_RGB2HSV)

image[:,:,2] = 100 # Cambiamos el valor de todos los valores V a 100
                   # que se encuentran en el índice 2 (0 -> H, 1 -> S, 2 -> V)

image2=cv2.cvtColor(image, cv2.COLOR_HSV2RGB)
plt.imshow(image2)#Solo se puede plotear en RGB 
```

```python
imggray= cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

imggray = cv2.cvtColor(imggray, cv2.COLOR_GRAY2RGB)
plt.imshow(imggray)

```

```python
image= cv2.cvtColor(img2, cv2.COLOR_RGB2HSV)

image[image[:,:,2]==255]=0 #seleccionamos aquellos que tengan el valor de 255 y lo volvemos en 0

image2=cv2.cvtColor(image, cv2.COLOR_HSV2RGB)

plt.imshow(image2)#Solo se puede plotear en RGB 
```

# Filtros

```python
plt.figure(figsize=(12,9))
img = cv2.imread('./files/brain.jpeg')
median = cv2.medianBlur(img, 5)
compare = np.concatenate((img, median), axis=1) 
plt.imshow(compare)
```

```python
plt.figure(figsize=(12,9))
gauss = cv2.GaussianBlur(img, (5,5), 0)
images = np.concatenate((median, gauss), axis=1)
plt.imshow(images)
```

```python
plt.figure(figsize=(12,9))

kernel = np.ones((5,5),np.float32)/25

print(kernel)

dst = cv2.filter2D(img,-1,kernel)#aplicando convolución


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Promediado')
plt.xticks([]), plt.yticks([])
plt.show()
```
