```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

# Algoritmo de Clustering Kmeans

```python
# Generar datos de ejemplo con 3 grupos
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
X[:, 1]
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Datos de ejemplo')
plt.show()
```

```python
# Instanciar el modelo K-Means con 3 clusters
kmeans = KMeans(n_clusters=3)

# Entrenar el modelo
kmeans.fit(X)

# Obtener las etiquetas de los clusters y los centros de los clusters
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
labels
```

```python
centers
```

```python
# Visualizar los puntos de datos con colores correspondientes a los clusters asignados por K-Means
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='inferno')

# Visualizar los centros de los clusters
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering con K-Means')
plt.show()
```

# Algoritmo de clustering DBSCAN

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
# Generar datos de ejemplo
X, _ = make_moons(n_samples=200, noise=0.1, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Datos de ejemplo')
plt.show()
```

```python
#intentamos con kmeans
kmeans = KMeans(n_clusters=2)

# Entrenar el modelo
kmeans.fit(X)

# Obtener las etiquetas de los clusters y los centros de los clusters
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Visualizar los puntos de datos con colores correspondientes a los clusters asignados por K-Means
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')

# Visualizar los centros de los clusters
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering con K-Means')
plt.show()
```

los grupos no quedan bien separados con el algoritmo kmeans por lo que se requiere uno más elaborado

```python
# Instanciar el modelo DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)

# Entrenar el modelo
dbscan.fit(X)

# Obtener las etiquetas de los clusters
labels = dbscan.labels_
```

```python
# Visualizar los puntos de datos con colores correspondientes a los clusters asignados por DBSCAN
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering con DBSCAN')
plt.show()
```


### Método de Clustering:

K-Means: Es un algoritmo de clustering particional, lo que significa que divide el conjunto de datos en un número predeterminado de clusters.

DBSCAN: Es un algoritmo de clustering basado en la densidad, que agrupa puntos en función de su densidad de vecindario.

### Número de clusters:

K-Means: Requiere que especifiques el número de clusters de antemano.
DBSCAN: No requiere especificar el número de clusters de antemano. Encuentra automáticamente el número de clusters basándose en la densidad de los datos y otros parámetros.

### Forma de los clusters:

K-Means: Asigna cada punto al cluster más cercano al centroide. Los clusters en K-Means tienden a ser de forma convexa y globalmente isotrópica (es decir, con igualdad de propiedades en todas las direcciones).
DBSCAN: Puede identificar clusters de cualquier forma y tamaño. Es capaz de detectar clusters de forma irregular y no necesariamente convexa.
Sensibilidad al ruido:

K-Means: Es sensible al ruido y a los valores atípicos, ya que asigna cada punto a un cluster incluso si está lejos de los centroides.

DBSCAN: Es robusto al ruido y puede manejar puntos atípicos como ruido, asignándolos a un cluster separado o etiquetándolos como puntos de ruido.

### Parámetros:

K-Means: Los principales parámetros son el número de clusters (n_clusters) y los centroides iniciales (init). También puede haber otros parámetros relacionados con la inicialización y la convergencia del algoritmo.

DBSCAN: Los principales parámetros son la distancia máxima entre puntos para considerarlos vecinos (eps) y el número mínimo de puntos en un vecindario para considerar un punto como núcleo (min_samples). Estos parámetros controlan la sensibilidad del algoritmo a la densidad y pueden variar dependiendo del conjunto de datos.

En resumen, mientras que K-Means es más adecuado para datos con clusters bien definidos y convexos, DBSCAN es más flexible y robusto, capaz de identificar clusters de cualquier forma y tamaño sin requerir el número de clusters de antemano y siendo menos sensible al ruido

```python

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generar datos de ejemplo
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizar los datos
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Datos de ejemplo')
plt.show()

# Instanciar el modelo de clustering aglomerativo
agglomerative = AgglomerativeClustering(n_clusters=None, distance_threshold=0)

# Entrenar el modelo
agglomerative.fit(X)

# Obtener las etiquetas de los clusters
labels = agglomerative.labels_

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering Aglomerativo')
plt.show()

# Calcular y visualizar el dendrograma
def plot_dendrogram(model, **kwargs):
    # Crear el dendrograma
    linkage_matrix = linkage(X, method='ward')  # Calcula la matriz de enlace utilizando el método de Ward

    # Representar el dendrograma
    dendrogram(linkage_matrix, **kwargs)

plt.title('Dendrograma de Clustering Aglomerativo')
plot_dendrogram(agglomerative)
plt.xlabel('Índice de muestra')
plt.ylabel('Distancia')
plt.show()
```
